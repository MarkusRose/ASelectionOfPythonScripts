```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
```

```python
mpl.colors.XKCD_COLORS
```

```python
xvals = np.linspace(0,1,256)
yvals = [0]*len(xvals)
print(len(mpl.colors.XKCD_COLORS)/3)
plt.figure(figsize=(3.3*2,3.3*30),dpi=80)
plt.plot(xvals,yvals,'-',lw=0)
yheight = 0
imcounter = 0
for colorname in mpl.colors.XKCD_COLORS:
    plt.axhspan(yheight,yheight+1,lw=0,color=colorname)
    plt.text(1.05,yheight+0.2,colorname,color='k')
    yheight += 1
    if yheight % 300 == 0:
        plt.xlim([0,1])
        plt.ylim([0,301])
        plt.tight_layout()
        plt.savefig("xkcdcolors{:}to{:}.png".format(imcounter*300,(imcounter+1)*300,dpi=80))
        plt.close()
        plt.figure(figsize=(3.3,3.3*30),dpi=80)
        plt.plot(xvals,yvals,'-',lw=0)
        imcounter += 1
        yheight = 0
plt.xlim([0,1])
plt.ylim([0,301])
plt.tight_layout()
plt.savefig("xkcdcolors900to949.png",dpi=80)
plt.close()
```
